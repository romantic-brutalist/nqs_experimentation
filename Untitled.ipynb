```python
import numpy as np
from scipy import sparse
from scipy.sparse.linalg import eigsh
```

```python
class ThreeSpinSystem:
    def __init__(self):
        # Pauli matrices
        self.sigma_x = sparse.csr_matrix([[0, 1], [1, 0]])
        self.sigma_y = sparse.csr_matrix([[0, -1j], [1j, 0]])
        self.sigma_z = sparse.csr_matrix([[1, 0], [0, -1]])
        self.I = sparse.eye(2)
        
        # Create full basis for 3 spins (8x8 matrices)
        self.basis_size = 2**3
    def get_pauli_operator(self, operator_type, position):
        """Creates a Pauli operator for a specific position in the 3-spin system"""
        if position == 0:
            if operator_type == 'x':
                return sparse.kron(sparse.kron(self.sigma_x, self.I), self.I)
            elif operator_type == 'y':
                return sparse.kron(sparse.kron(self.sigma_y, self.I), self.I)
            elif operator_type == 'z':
                return sparse.kron(sparse.kron(self.sigma_z, self.I), self.I)
        elif position == 1:
            if operator_type == 'x':
                return sparse.kron(sparse.kron(self.I, self.sigma_x), self.I)
            elif operator_type == 'y':
                return sparse.kron(sparse.kron(self.I, self.sigma_y), self.I)
            elif operator_type == 'z':
                return sparse.kron(sparse.kron(self.I, self.sigma_z), self.I)
        elif position == 2:
            if operator_type == 'x':
                return sparse.kron(sparse.kron(self.I, self.I), self.sigma_x)
            elif operator_type == 'y':
                return sparse.kron(sparse.kron(self.I, self.I), self.sigma_y)
            elif operator_type == 'z':
                return sparse.kron(sparse.kron(self.I, self.I), self.sigma_z)
        
    def create_ising_hamiltonian(self, h):
        """Creates transverse field Ising Hamiltonian: -h∑σx - ∑σz_iσz_j"""
        H = sparse.csr_matrix((self.basis_size, self.basis_size))
        
        # Transverse field terms
        for i in range(3):
            H -= h * self.get_pauli_operator('x', i)
            
        # Interaction terms
        H -= (self.get_pauli_operator('z', 0) @ self.get_pauli_operator('z', 1) +
              self.get_pauli_operator('z', 1) @ self.get_pauli_operator('z', 2) +
              self.get_pauli_operator('z', 2) @ self.get_pauli_operator('z', 0))
              
        return H

    def find_ground_state(self, hamiltonian):
        """Finds the ground state energy and wavefunction using sparse eigendecomposition"""
        eigenvalues, eigenvectors = eigsh(hamiltonian, k=1, which='SA')
        return eigenvalues[0], eigenvectors[:, 0]
                
```

```python
import numpy as np

def random_complex(size):
    """Initialize small random complex numbers for neural network parameters."""
    a = (np.random.random(size) - .5) * 1e-2
    b = (np.random.random(size) - .5) * 1e-2
    return a + 1j*b

class NeuralQuantumState:
    def __init__(self, n_spins, alpha):
        """Initialize the Neural Quantum State with the RBM architecture.
        
        Args:
            n_spins: Number of physical spins (visible neurons)
            alpha: Ratio of hidden to visible neurons (hidden layer density)
        """
        self.n_spins = n_spins
        self.n_hidden = alpha * n_spins
        self.state_indices = list(range(n_spins))  # Store indices for sampling
        
        # Initialize network parameters with small random values
        self.a = random_complex(n_spins)
        self.b = random_complex(self.n_hidden)
        self.W = random_complex((n_spins, self.n_hidden))

    def effective_angles(self, state):
        """Calculate the effective angles θⱼ(S) = bⱼ + ∑ᵢ Wᵢⱼsᵢ."""
        return self.b + np.inner(np.transpose(self.W), state)

    def psi(self, state):
        """Calculate wave function amplitude Ψ(S) = exp(∑ᵢaᵢsᵢ) ∏ⱼ 2cosh(θⱼ)."""
        return np.exp(np.inner(self.a, state)) * np.prod(2*np.cosh(self.effective_angles(state)))

    def local_energy(self, state, h):
        """Calculate the local energy E_loc(S) = ⟨S|H|Ψ⟩/Ψ(S)."""
        E = 0
        # Interaction terms (σᶻᵢσᶻⱼ)
        for i in range(self.n_spins):
            if i == self.n_spins-1:
                E += (state[i]*state[0])
            else:
                E -= (state[i]*state[i+1])

        # Transverse field terms (σˣ)
        psi_s = self.psi(state)
        for i in range(self.n_spins):
            state[i] *= -1  # Flip spin i
            E += h * self.psi(state)/psi_s
            state[i] *= -1  # Flip back
            
        return E/self.n_spins

    def metropolis_step(self, state, h):
        """Perform one Metropolis-Hastings step for the given state."""
        # Choose random spin to flip
        sites = np.random.choice(self.state_indices, 1)
        psi_before = self.psi(state)
        
        # Flip the chosen spin
        for i in sites:
            state[i] *= -1
        
        # Calculate acceptance probability
        psi_after = self.psi(state)
        acceptance = np.real(psi_after*np.conj(psi_after)/(psi_before*np.conj(psi_before)))
        
        # Accept or reject the move
        if acceptance < np.random.uniform():
            for i in sites:
                state[i] *= -1  # Restore original state if rejected
            return state, 1
        return state, 0

def train_nqs(n_spins=3, alpha=2, h=2.0, n_blocks=30):
    """Train the Neural Quantum State to find the ground state energy.
    
    Args:
        n_spins: Number of spins in the system
        alpha: Hidden layer density
        h: Transverse field strength
        n_blocks: Number of training iterations
        
    Returns:
        list: Energy values during training
    """
    nqs = NeuralQuantumState(n_spins, alpha)
    block_energies = []
    
    for block_i in range(n_blocks):
        # Initialize random state for this block
        state = np.random.randint(2, size=n_spins)
        state[state == 0] = -1
        
        # Thermalization
        for _ in range(10000):
            state, _ = nqs.metropolis_step(state, h)
        
        # Monte Carlo sampling
        iterations = 20000
        rejected = 0
        array_E_loc = []
        array_a_d = []
        array_b_d = []
        array_w_d = []
        
        for k in range(iterations):
            state, was_rejected = nqs.metropolis_step(state, h)
            rejected += was_rejected
            
            if k % 100 == 0:
                # Calculate local energy
                E_loc = nqs.local_energy(state, h)
                
                # Calculate gradients
                psi_s = nqs.psi(state)
                
                # Gradient for a parameters
                a_deriv = np.zeros(n_spins, dtype=np.complex_)
                for i in range(n_spins):
                    state[i] *= -1  # flip
                    a_deriv[i] = -h*nqs.psi(state)/psi_s*2.*state[i]
                    state[i] *= -1  # flip back
                
                # Gradient for b parameters
                b_deriv = np.zeros(nqs.n_hidden, dtype=np.complex_)
                for b_j in range(nqs.n_hidden):
                    theta_before = nqs.effective_angles(state)
                    db_sum = 0
                    for i in range(n_spins):
                        state[i] *= -1
                        theta_after = nqs.effective_angles(state)
                        db_sum += nqs.psi(state)/psi_s * (
                            np.tanh(theta_after[b_j]) - 
                            np.tanh(theta_before[b_j])
                        )
                        state[i] *= -1
                    b_deriv[b_j] = -h * db_sum
                
                # Gradient for W parameters
                W_deriv = np.zeros((n_spins, nqs.n_hidden), dtype=np.complex_)
                for w_i in range(n_spins):
                    for w_j in range(nqs.n_hidden):
                        theta_before = nqs.effective_angles(state)
                        dw_sum = 0
                        for i in range(n_spins):
                            state[i] *= -1
                            theta_after = nqs.effective_angles(state)
                            dw_sum += nqs.psi(state)/psi_s * (
                                -state[i]*np.tanh(theta_after[w_j]) -
                                state[i]*np.tanh(theta_before[w_j])
                            )
                            state[i] *= -1
                        W_deriv[w_i, w_j] = h * dw_sum
                
                # Store measurements
                array_E_loc.append(np.real(E_loc))
                array_a_d.append(a_deriv)
                array_b_d.append(b_deriv)
                array_w_d.append(W_deriv)
        
        # Update parameters
        mean_E = np.mean(array_E_loc)
        std_E = np.std(array_E_loc)/np.sqrt(len(array_E_loc))
        rejection_rate = 100.0 * rejected/iterations
        
        # Calculate mean gradients and update parameters
        mean_da = np.mean(np.array(array_a_d), axis=0)
        mean_db = np.mean(np.array(array_b_d), axis=0)
        mean_dw = np.mean(np.array(array_w_d), axis=0)
        
        nqs.a = nqs.a - 0.1 * mean_da
        nqs.b = nqs.b - 0.1 * mean_db
        nqs.W = nqs.W - 0.1 * mean_dw
        
        print(f'{block_i+1}. E_loc={mean_E:.4f} std={std_E:.4f} '
              f'({rejection_rate:.1f}% moves rejected)')
        
        block_energies.append(mean_E)
    
    return block_energies,nqs

# Run training
if __name__ == "__main__":
    energies,nqs = train_nqs()
    
    # Plot results
    import matplotlib.pyplot as plt
    plt.figure(figsize=(10, 6))
    plt.plot(energies)
    plt.xlabel('Training Block')
    plt.ylabel('Energy per Spin')
    plt.title('Neural Quantum State Training')
    plt.grid(True)
    plt.show()
```

```python
from itertools import product
import numpy as np

def calculate_full_wavefunction(nqs):
    """
    Calculate the wave function for all possible basis states using the trained NQS.
    
    Args:
        nqs: Trained NeuralQuantumState object
    
    Returns:
        states: List of all possible basis states
        psi: Corresponding wave function amplitudes
        probabilities: Probability of finding the system in each state
    """
    # Generate all possible basis states
    basis_states = list(product([0, 1], repeat=nqs.n_spins))
    
    # Calculate psi for each state
    psi = np.zeros(len(basis_states), dtype=np.complex128)
    for i, state in enumerate(basis_states):
        psi[i] = nqs.psi(np.array(state))
    
    # Normalize the wave function
    normalization = np.sqrt(np.sum(np.abs(psi)**2))
    psi = psi / normalization
    
    # Calculate probabilities
    probabilities = np.abs(psi)**2
    
    return basis_states, psi, probabilities

bs,psi,probs = calculate_full_wavefunction(nqs)
```

```python
def print_wavefunction(basis_states, psi, probabilities, threshold=1e-3):
    """
    Print the wave function in a readable format, showing only significant contributions.
    
    Args:
        basis_states: List of all possible basis states
        psi: Wave function amplitudes
        probabilities: Probability of finding the system in each state
        threshold: Minimum probability to display
    """
    print("\nSignificant components of the wave function:")
    print("State | Amplitude (real, imag) | Probability")
    print("-" * 45)
    
    for state, amp, prob in zip(basis_states, psi, probabilities):
        if prob > threshold:
            state_str = ''.join(['1' if s == 1 else '0' for s in state])
            print(f"{state_str} | {amp.real:6.3f}, {amp.imag:6.3f}i | {prob:6.3f}")
    
    # Verify normalization
    total_prob = np.sum(probabilities)
    print(f"\nTotal probability: {total_prob:.6f}")
```

```python
import numpy as np
from itertools import product
import matplotlib.pyplot as plt

def compare_nqs_with_exact(n_spins=3, alpha=2, h=2.0, n_blocks=30):
    """
    Compare the Neural Quantum State approach with exact diagonalization.
    This function serves as a comprehensive test of our implementation by:
    1. Training the NQS to find the ground state
    2. Computing the exact ground state through diagonalization
    3. Comparing the results and visualizing the convergence
    """
    print("Starting Neural Quantum State training...\n")
    nqs = NeuralQuantumState(n_spins, alpha)
    training_energies = []
    
    # Initial state for NQS training
    state = np.random.randint(2, size=n_spins)
    state[state == 0] = -1
    
    # Training loop
    for block_i in range(n_blocks):
        # Thermalization
        for _ in range(10000):
            state, _ = nqs.metropolis_step(state, h)
        
        # Monte Carlo sampling
        iterations = 20000
        rejected = 0
        array_E_loc = []
        array_a_d = []
        array_b_d = []
        array_w_d = []
        
        for k in range(iterations):
            state, was_rejected = nqs.metropolis_step(state, h)
            rejected += was_rejected
            
            if k % 100 == 0:  # Take measurements every 100 steps
                # Calculate local energy and gradients
                E_loc = nqs.local_energy(state, h)
                psi_s = nqs.psi(state)
                
                # Compute gradients for each parameter type
                # Visible bias gradients
                a_deriv = np.zeros(n_spins, dtype=np.complex_)
                for i in range(n_spins):
                    state[i] *= -1
                    a_deriv[i] = -h * nqs.psi(state)/psi_s * 2 * state[i]
                    state[i] *= -1
                
                # Hidden bias gradients
                b_deriv = np.zeros(nqs.n_hidden, dtype=np.complex_)
                for b_j in range(nqs.n_hidden):
                    before_flip = np.tanh(nqs.effective_angles(state))
                    db_sum = 0
                    for i in range(n_spins):
                        state[i] *= -1
                        db_sum += nqs.psi(state)/psi_s * (
                            np.tanh(nqs.effective_angles(state)[b_j]) - before_flip[b_j]
                        )
                        state[i] *= -1
                    b_deriv[b_j] = -h * db_sum
                
                # Weight gradients
                W_deriv = np.zeros((n_spins, nqs.n_hidden), dtype=np.complex_)
                for w_i in range(n_spins):
                    for w_j in range(nqs.n_hidden):
                        before_flip = np.tanh(nqs.effective_angles(state))
                        dw_sum = 0
                        for i in range(n_spins):
                            state[i] *= -1
                            dw_sum += nqs.psi(state)/psi_s * (
                                -state[i] * np.tanh(nqs.effective_angles(state)[w_j]) -
                                state[i] * before_flip[w_j]
                            )
                            state[i] *= -1
                        W_deriv[w_i, w_j] = h * dw_sum
                
                # Store measurements
                array_E_loc.append(np.real(E_loc))
                array_a_d.append(a_deriv)
                array_b_d.append(b_deriv)
                array_w_d.append(W_deriv)
        
        # Update parameters using gradient descent
        mean_E = np.mean(array_E_loc)
        mean_da = np.mean(np.array(array_a_d), axis=0)
        mean_db = np.mean(np.array(array_b_d), axis=0)
        mean_dw = np.mean(np.array(array_w_d), axis=0)
        
        nqs.a = nqs.a - 0.1 * mean_da
        nqs.b = nqs.b - 0.1 * mean_db
        nqs.W = nqs.W - 0.1 * mean_dw
        
        # Calculate statistics
        std_E = np.std(array_E_loc)/np.sqrt(len(array_E_loc))
        rejection_rate = 100.0 * rejected/iterations
        
        print(f'{block_i+1}. E_loc={mean_E:.4f} std={std_E:.4f} '
              f'({rejection_rate:.1f}% moves rejected)')
        
        training_energies.append(mean_E)
    
    print("\nCalculating exact solution...")
    # Generate all possible basis states
    basis = list(product([-1, 1], repeat=n_spins))
    H = np.zeros((2**n_spins, 2**n_spins))
    
    # Construct the Hamiltonian matrix
    for H_i in range(2**n_spins):
        for H_j in range(2**n_spins):
            H_sum = 0
            # Interaction terms
            for i in range(n_spins):
                if H_i == H_j:
                    if i == n_spins-1:
                        H_sum += basis[H_j][i] * basis[H_j][0]
                    else:
                        H_sum += basis[H_j][i] * basis[H_j][i+1]
            
            # Transverse field terms
            for i in range(n_spins):
                sj = list(basis[H_j])
                sj[i] *= -1
                if H_i == basis.index(tuple(sj)):
                    H_sum += h
            
            H[H_i, H_j] = H_sum
    
    # Find ground state energy through diagonalization
    exact_energy = np.min(np.real(np.linalg.eigvals(H)))/n_spins
    
    # Plot comparison
    plt.figure(figsize=(10, 6))
    plt.plot(training_energies, 'b-', label='NQS Energy')
    plt.axhline(y=exact_energy, color='r', linestyle='--', 
                label=f'Exact Energy: {exact_energy:.4f}')
    plt.xlabel('Training Block')
    plt.ylabel('Energy per Spin')
    plt.title('Neural Quantum State Training vs Exact Ground State')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Print final comparison
    final_nqs_energy = training_energies[-1]
    error = abs(final_nqs_energy - exact_energy)
    print(f'\nFinal Results:')
    print(f'NQS Energy: {final_nqs_energy:.6f}')
    print(f'Exact Energy: {exact_energy:.6f}')
    print(f'Absolute Error: {error:.6f}')
    print(f'Relative Error: {100*error/abs(exact_energy):.2f}%')

# Run comparison if script is executed directly
if __name__ == "__main__":
    compare_nqs_with_exact(n_spins=3, alpha=1, h=2.0, n_blocks=10)
```
